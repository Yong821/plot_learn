```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = np.arange(10)
fig, ax = plt.subplots(2, 2)
ax[0][0].plot(x, np.exp(x))
ax[0][0].set_xlabel("plot1")
ax[0][0].set_ylabel("plot1")
ax[0][1].plot(x, np.sqrt(x))
ax[0][1].set_xlabel("plot1")
ax[0][1].set_ylabel("plot1")
ax[1][0].plot(x, x * x)
ax[1][0].set_xlabel("plot1")
ax[1][1].plot(x, x * x * x)
```
