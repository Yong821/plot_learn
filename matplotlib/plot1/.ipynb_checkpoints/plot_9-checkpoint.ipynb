```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(1, 11, 1)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(x, np.exp(x), c='b')
ax[0].plot(x, x ** 2, c='g')
ax[0].set_title('no log_y_scale')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].plot(x, np.exp(x), c='b')
ax[1].plot(x, x ** 2, c='g')
ax[1].set_title('with log_y_scale')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_yscale('log')
```
