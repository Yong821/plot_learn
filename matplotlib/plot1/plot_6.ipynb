```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.5, 0.5, 0.5, 0.5])
ax1.plot(x, y1, '--')
ax2.plot(x, y2, '.')
ax1.set_title("fig1")
ax2.set_title("fig2")
ax1.set_xlabel("param1")
ax1.set_ylabel("y1")
ax2.set_xlabel("param2")
ax2.set_ylabel("y2")
```
