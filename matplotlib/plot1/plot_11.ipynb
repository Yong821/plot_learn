```python
    import matplotlib.pyplot as plt
    import numpy as np
    import math
    x = np.arange(0, math.pi*2, 0.05)
    #生成画布对象
    fig = plt.figure()
    #添加绘图区域
    ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    y = np.sin(x)
    ax.plot(x, y)
    #设置x轴标签
    ax.set_xlabel('angle')
    ax.set_title('sine')
    #设置x轴刻度以及标签
#     ax.set_xticklabels(['zero','two','four','six'])
    ax.set_xticks([0,2,4,6], ['zero', 'two', 'four', 'six'])
    #设置y轴刻度
    ax.set_yticks([-1,0,1])
    plt.show()
```
