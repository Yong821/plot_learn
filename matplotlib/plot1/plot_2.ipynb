```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(0, 2, 0.05)
y = np.sin(x)
```

```python
# 创建gifure对象
fig = plt.figure(figsize=(10, 10), frameon=False)
# 添加轴
ax = fig.add_axes([0, 0.8, 1, 1])
ax.set_title("sine wave")
ax.set_xlabel("angle")
ax.set_ylabel("sine")
ax.plot(x, y)
```
