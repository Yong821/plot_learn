```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 20
x1 = np.random.normal(0, 1, n)
x2 = np.random.normal(3, 4, n)
y = np.random.uniform(-10, 10, n)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x1, y, 'ys--')
ax.plot(x2, y, 'gs--')
ax.legend(labels=['param1', 'param2'])
ax.set_xlabel("mideum")
ax.set_ylabel("sale")
ax.set_title('wine')
```
