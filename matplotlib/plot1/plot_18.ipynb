```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X ** 2 + Y ** 2)
fig, ax = plt.subplots()
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp)
ax.set_title("Filled Contours Plot")
ax.set_xlabel('x (cm)')
ax.set_ylabel('x (cm)')
ax.contour(X, Y, Z)
```
