```python
import matplotlib.pyplot as plt
```

```python
x = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y = [12, 13, 21, 6, 12, 32, 20]
fig, ax = plt.subplots()
ax.set_ylabel("weekDay")
ax.set_xlabel("Day")
ax.set_title("hot day")
ax.legend("week hot day")
ax.plot(x, y)

for x1, y1 in zip(x, y):
    ax.text(x1, y1, str(y1), ha='center', fontsize=13)
```
